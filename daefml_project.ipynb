```python
fig = px.density_heatmap(df, x="year", y="month",title='Violence against Women Yearly')
fig.show()
```

```python
#Save for next time since we have only ONE month
import numpy as np

df_year_month_day = df.groupby(by=['year','month','day']).count().survey

```

```python
df["day"]
```

```python
df_year_month_day = df_year_month_day.reset_index().pivot('year','month','day').transpose()

fig = px.line(np.cumsum(df_year_month_day), title='Number of Violence against Women Cumulative',markers=True)
fig.show()
```

```python
question_df =df.groupby(["Question"])["Value"].agg(["median", "max", "min", "mean"]).reset_index()
question_df.columns = ["Question", "Median", "Max", "Min", "Mean"]
print("Violence % median, min, max, and mean per question asked")
question_df.sort_values(["Median"])
```

```python
demoq_df = df.groupby(["demoQuestions", "demoResponses", "Question"])["Value"].agg(["median", "max", "min", "mean"]).reset_index()
demoq_df.columns = ["demoQuestions", "demoResponses", "Question", "Median", "Max", "Min", "Mean"]
print("Violence % median, min, max, and mean per demographic group and question asked")
demoq_df = demoq_df.sort_values(["demoQuestions", "demoResponses", "Median"])
demoq_df.head()
```
